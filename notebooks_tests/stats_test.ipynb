```python
import pandas as pd
import statistics
import stats

# File paths
RS_LC_FS = "./results/lines_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_random_search.csv"
RS_LC_GS = "./results/lines_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_random_search_new_diss.csv"
RS_SC_FS = "./results/spectr_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_random_search.csv"
RS_SC_GS = "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_random_search_new_diss.csv"

GA_LC_FS = "./results/lines_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean.csv"
GA_AGRM_LC_FS = "./results/lines_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_aggr_mut.csv"
GA_150_25_3_LC_FS = "./results/lines_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_150_25_3.csv"

GA_LC_GS = "./results/lines_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_new_diss.csv"
GA_AGRM_LC_GS = "./results/lines_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_aggr_mut_new_diss.csv"
GA_150_25_3_LC_GS = "./results/lines_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_150_25_3_new_diss.csv"

GA_SC_FS = "./results/spectr_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit.csv"
GA_AGRM_SC_FS = "./results/spectr_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_aggr_mut.csv"
GA_150_25_3_SC_FS = "./results/spectr_comp_fitness/first_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_150_25_3.csv"

GA_SC_GS = "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_new_diss.csv"
GA_AGRM_SC_GS = "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_aggr_mut_new_diss.csv"
GA_150_25_3_SC_GS = "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_150_25_3_new_diss.csv"

# Define all file comparisons
file_pairs = [
    # LC_FS group
    (RS_LC_FS, GA_LC_FS, "RS_LC_FS vs GA_LC_FS"),
    (RS_LC_FS, GA_AGRM_LC_FS, "RS_LC_FS vs GA_AGRM_LC_FS"),
    (RS_LC_FS, GA_150_25_3_LC_FS, "RS_LC_FS vs GA_150_25_3_LC_FS"),

    # LC_GS group
    (RS_LC_GS, GA_LC_GS, "RS_LC_GS vs GA_LC_GS"),
    (RS_LC_GS, GA_AGRM_LC_GS, "RS_LC_GS vs GA_AGRM_LC_GS"),
    (RS_LC_GS, GA_150_25_3_LC_GS, "RS_LC_GS vs GA_150_25_3_LC_GS"),

    # SC_FS group
    (RS_SC_FS, GA_SC_FS, "RS_SC_FS vs GA_SC_FS"),
    (RS_SC_FS, GA_AGRM_SC_FS, "RS_SC_FS vs GA_AGRM_SC_FS"),
    (RS_SC_FS, GA_150_25_3_SC_FS, "RS_SC_FS vs GA_150_25_3_SC_FS"),

    # SC_GS group
    (RS_SC_GS, GA_SC_GS, "RS_SC_GS vs GA_SC_GS"),
    (RS_SC_GS, GA_AGRM_SC_GS, "RS_SC_GS vs GA_AGRM_SC_GS"),
    (RS_SC_GS, GA_150_25_3_SC_GS, "RS_SC_GS vs GA_150_25_3_SC_GS")
]

# Loop through each file pair
for file1, file2, label in file_pairs:
    # Load data from file1
    df1 = pd.read_csv(file1)
    similarity_indivd_file1 = df1['similarity_indv'].astype(float).tolist()
    variance_rs = statistics.variance(similarity_indivd_file1)
    mean_rs = statistics.median(similarity_indivd_file1)
    print(f"{label} - Variance RS: {variance_rs}, Median RS: {mean_rs}")
    
    # Load data from file2
    df2 = pd.read_csv(file2)
    similarity_indivd_file2 = df2['similarity_indv'].astype(float).tolist()
    variance_ga = statistics.variance(similarity_indivd_file2)
    mean_ga = statistics.median(similarity_indivd_file2)
    print(f"{label} - Variance GA: {variance_ga}, Median GA: {mean_ga}")
    
    # Perform Mann-Whitney U test using custom stats function
    is_significant, p_value = stats.is_mannwhitneyu_passed(similarity_indivd_file2, similarity_indivd_file1, alternative="greater")
    print(f"{label} - Mann-Whitney U test result: Significant={is_significant}, p-value={p_value}")
    print()
```

```python
import stats
import pandas as pd

# File paths and method names
file_paths = {
    "Random Search": "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_random_search_new_diss.csv",
    "GA w Clean": "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_new_diss.csv",
    "GA w Model": "./results/spectr_comp_fitness/gigi_sim/w_model/dataset_audios_guitar_low_ranges_w_model_spectr_fit_new_diss.csv",
    "GA aggr. mut.": "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_aggr_mut_new_diss.csv",
    "GA 150 25 3": "./results/spectr_comp_fitness/gigi_sim/w_clean/dataset_audios_guitar_low_ranges_w_clean_spectr_fit_150_25_3_new_diss.csv"
}

# Dictionary to hold method names and their corresponding list of float values
similarity_data = {}

# Load each file and populate the dictionary
for method_name, file_path in file_paths.items():
    df = pd.read_csv(file_path)
    similarity_data[method_name] = df['similarity_indv'].astype(float).tolist()
    
# Perform the Mann-Whitney U test
res = stats.perform_mannwhitneyu_holm_bonferroni(data=similarity_data, alternative="greater")

# Loop through each result and format the output
for result in res:
    if isinstance(result, tuple):
        # Print simple tuple results (e.g., a single test result)
        print("Simple Result:")
        print(f"  Passed: {result[0]}")
        print(f"  P-value: {result[1]:.6e}")
        print()  # Add a blank line for spacing

    elif isinstance(result, dict):
        # Check if itâ€™s a dictionary of comparisons with a boolean result
        if all(isinstance(val, bool) for val in result.values()):
            print("Comparison Results by Method:")
            for method, passed in result.items():
                print(f"  {method}: {'Passed' if passed else 'Failed'}")
            print()  # Add a blank line for spacing
        
        else:
            # Nested dictionary case, showing pairwise comparison results
            print("Pairwise Holm-Bonferroni Results:")
            for method, comparisons in result.items():
                print(f"Method: {method}")
                for comp_method, passed in comparisons.items():
                    print(f"  Compared with {comp_method}: {'Passed' if passed else 'Failed'}")
                print()  # Add a blank line for spacing

```
